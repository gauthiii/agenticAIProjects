```python
from dotenv import load_dotenv

load_dotenv()

import aisuite as ai

# Define the client. You can use this variable inside the functions!
CLIENT = ai.Client()


import PyPDF2

import re
import json

import sys
import io
```

```python
pdf_file_path = "7.pdf"
figma_report = ""

with open(pdf_file_path, "rb") as file:
    reader = PyPDF2.PdfReader(file)
    for page in reader.pages:
        figma_report += page.extract_text()

print(figma_report)
```

```python
from docx import Document

# Open the document
document = Document('EPIC Guidelines.docx')

# Iterate through each paragraph and print its text
for paragraph in document.paragraphs:
    print(paragraph.text)

# To get all text in a single string
epic_guidelines = ""
for paragraph in document.paragraphs:
    epic_guidelines += paragraph.text + "\n"
print(epic_guidelines)
```

```python
print(len(epic_guidelines))
print(len(epic_guidelines.split(" ")))
```

```python
def planner_ollama(query, figma_report, epic_guidelines, model: str = "ollama:gemma3:latest") -> str:
    """
    Use a local Ollama model as a planner.
    """



    prompt = f"""

    The user has a task: {query}

    The source documentation is a Figma Frame Export: {figma_report}

    The guidelines for creating the EPIC documentation is the following: {epic_guidelines}


    Your job is to:
    1. Think step by step about how you would solve the task.
    2. Output a clear, numbered list of steps to execute this plan.

    Important:
    - Do NOT add extra commentary; just the steps.
    """

    response = CLIENT.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": "You are a senior business analyst who is strong at agile methodology."},
            {"role": "user", "content": prompt},
        ],
        temperature=0.3,
    )

    return response.choices[0].message.content
    
```

```python
res = planner_ollama("Draft an EPIC documentation for the following source documentation",figma_report,epic_guidelines)
```

```python
print(res)
```

```python
def executor(strategy, figma_report, epic_guidelines, model: str = "ollama:gemma3:latest") -> str:
    """
    Use a local Ollama model as an executor.
    """



    prompt = f"""

    The user has a task with a strategy: {strategy}

    The source documentation is a Figma Frame Export: {figma_report}

    The guidelines for creating the EPIC documentation is the following: {epic_guidelines}


    Your job is to:
    1. Execute the strategy and give me the final EPIC documentation.

    Important:
    - Do NOT add extra commentary; just the steps.
    """

    response = CLIENT.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": "You are a senior business analyst who is strong at agile methodology."},
            {"role": "user", "content": prompt},
        ],
        temperature=0.3,
    )

    return response.choices[0].message.content
```

```python
doc = executor(res,figma_report,epic_guidelines)
```
