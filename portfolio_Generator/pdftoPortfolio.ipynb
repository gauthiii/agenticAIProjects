````python
from dotenv import load_dotenv

load_dotenv()

import aisuite as ai

# Define the client. You can use this variable inside the functions!
CLIENT = ai.Client()


import PyPDF2

import re
import json

import sys
import io
````

````python
pdf_file_path = "resume.pdf"
text = ""

with open(pdf_file_path, "rb") as file:
    reader = PyPDF2.PdfReader(file)
    for page in reader.pages:
        text += page.extract_text()

print(text)
````

````python
# Agent1: Basic Details agent

def parse_basic_details(text: str, model: str = "openai:gpt-4o") -> str: 
    
    ### START CODE HERE ###

    # Define your prompt here. A multi-line f-string is typically used for this.
    prompt = f'''
    
    You are a resume parser.
    
    The details of the resume parsed from a pdf is here: {text}
    
    You need to extract information and store them in the form of a dict.

    The dict should have the following keys:

    - name
    - email
    - phone
    - linkedinUrl
    - socialUrls (other than linkedIn. It can be anything. If there are mutiple store them as a dict with the platform name as key and ur as value. If none, store as an empty dict)
    - educationDetails (store as a list of dicts with keys: degree, major, institution, yearOfCompetion, grade or percentage if available. if empty, store as an empty list)
    - workExperience (store as a list of dict with keys: companyName, role, startDate, endDate, duration, responsibilities (store as a list of 3 concise points). if empty, store as an empty list)
    - projects (store as a list of dict with keys: projectName, description, tech stack or platforms used. (store as a list). if empty, store as an empty list)
    - skills (categories them first. And use the following categories as keys for a dict with each of them having the list of the skills as values. Have 4 categories maximum)
    - summary (use the summary or bio or objective section. If not available, generate a concise summary in 2-3 sentences based on everything else available)

    Return only the dict. No explanations.
    
    ''' 

    ### END CODE HERE ###
    
    # Get a response from the LLM by creating a chat with the client.
    response = CLIENT.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=1.0,
    )

    return response.choices[0].message.content
````

````python
# Agent2: Web Skeleton

def web_skeleton_generator(resume_obj_final: dict, model: str = "anthropic:claude-haiku-4-5") -> str: 
    
    ### START CODE HERE ###

    # Define your prompt here. A multi-line f-string is typically used for this.
    prompt = f'''
    
    You have the details of the resume parsed right here: {resume_obj_final}
    
    Produce a pixel-close, accessible, single-page portfolio 
    site using vanilla HTML + Tailwind CDN + small JS modules 
    that exactly matches the structure/behaviors below. No build tools. 
    
    Tech & Constraints:

    - No frameworks (no React/Next). Use Tailwind via CDN.
    - Include Font Awesome kit tag: <script src="https://kit.fontawesome.com/dee65e5b65.js" crossorigin="anonymous"></script>
    - Support dark mode using class="dark" on <html> and persist with localStorage.
    - Smooth scrolling, sticky nav, scroll progress bar, reveal-on-scroll animations (reduced-motion aware).
    - Mobile menu (toggle visibility), basic scrollspy highlighting active section link.
    
    BUT DO NOT GIVE THE WHOLE WBESITE, only the SKELETON of the WEBISTE.

    In other words. Just give the basic website structure with Tailwind classes.
    Sections and but no content. 
    
    Include a the background, the nav bar with the follwowing links:

    - About
    - Skills
    - Experience
    - Projects
    - Education
    - Contact
    - Theme Toggle

    After that just create a section for each of the above links.

    But no content inside the sections

    Only the title of the html page with "Name - Portfolio" 
    and basic skeleton using the tailwind css and js for the above features.

    

    Return only the html code. No explanations.
    
    ''' 

    ### END CODE HERE ###
    
    # Get a response from the LLM by creating a chat with the client.
    response = CLIENT.chat.completions.create(
        model=model,
        messages=[

            {"role": "system", "content": "You are a senior frontend engineer."},
            {"role": "user", "content": prompt}
            ],
        temperature=1.0,
    )

    return response.choices[0].message.content
````

````python
# Agent 2
skeleton = web_skeleton_generator(resume_obj_final)
````

````python
import re
from pathlib import Path

def save_fenced_html(s: str, out_path: str | Path) -> Path:
    """
    Extracts HTML from a fenced block like:
    ```html
    ...html here...
    ```
    and saves it to out_path.
    If no fence is found, saves s as-is.
    """
    # Match ```html ... ``` OR ```HTML ... ```
    m = re.search(r"^```html\s*(.*?)\s*```$", s, flags=re.DOTALL | re.IGNORECASE)
    html = m.group(1) if m else s
    out_path = Path(out_path)
    out_path.write_text(html, encoding="utf-8")
    return out_path



save_fenced_html(skeleton, "skeleton.html")
print("Saved to skeleton.html")
````

````python
# Agent3: About Section

def about_generator(skeleton: str, resume_obj_final: dict, model: str = "anthropic:claude-haiku-4-5") -> str: 
    
    ### START CODE HERE ###

    # Define your prompt here. A multi-line f-string is typically used for this.
    prompt = f'''
    
    You have the details of the resume parsed right here
    with the variable name 'resume_obj_final': {resume_obj_final}
    
    You have the skeleton of the portfolio site 
    right here with the variable name 'skeleton': {skeleton}

    Now give me a the complete html section for the about section from the resume details.

    Since there's already the section defined, check if the section name "About" already exists in the code
    if it is already do not repeat the section name like 'About' in the content again.

    But make sure you give the python code to add that section to the 'skeleton' variable.
    So what that python code does it return the updated skeleton variable with about section added in the right place.

    Give me the deliverable as an executable python code like this:


    <execute_python>
    ...
    </execute_python>
    
    

    

    Return only in the above format. Nothing else.
    
    ''' 

    ### END CODE HERE ###
    
    # Get a response from the LLM by creating a chat with the client.
    response = CLIENT.chat.completions.create(
        model=model,
        messages=[

            {"role": "system", "content": "You are a senior frontend engineer."},
            {"role": "user", "content": prompt}
            ],
        temperature=1.0,
    )

    return response.choices[0].message.content
````

````python
# Agent 3
about_code = about_generator(skeleton,resume_obj_final)
````

````python
match = re.search(r"<execute_python>([\s\S]*?)</execute_python>", about_code)


if match:
    reflected_code = match.group(1).strip()
    output_capture = io.StringIO()
    sys.stdout = output_capture
    exec(reflected_code)

    sys.stdout = sys.__stdout__  # restore normal stdout
    skeleton2 = output_capture.getvalue()  # this contains printed output

    
    

save_fenced_html(skeleton2, "about.html")
print("Saved to about.html")
````
